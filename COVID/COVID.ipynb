```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_excel('dataset.xlsx')
```

```python
dataset.head(50)
```

```python
y=dataset['SARS-Cov-2 exam result']
columnsNotToDrop=['SARS-Cov-2 exam result']
g=dataset.columns
datatype=dataset.dtypes
for i in g[6:]:
    ser=dataset[i]
    ser.dropna(inplace=True)
    print(i,'\n',ser.size)
    if ser.size>=500:
        if datatype[i]=='int64' or datatype[i]=='float64':
            columnsNotToDrop.append(i)
```

```python
dataset=dataset[columnsNotToDrop]
```

```python
dataset.head(10)
```

```python
import seaborn as sns
```

```python
sns.countplot(dataset['SARS-Cov-2 exam result'])
```

```python
for col in dataset.columns[1:]:
    sns.swarmplot(data=dataset,x=col,y='SARS-Cov-2 exam result')
    plt.show()
```

```python
dataset=dataset.fillna(dataset.mean())
```

```python
dataset.head(10)
```

```python
from sklearn.model_selection import train_test_split
y=dataset['SARS-Cov-2 exam result']
y=y.map({'positive':1,'negative':0})
```

```python
dataset.head(5)
```

```python
columnsNotToDrop.pop(0)
dataset[columnsNotToDrop]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset[columnsNotToDrop],y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(class_weight="balanced")
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,lr_pred))
```

```python
print(classification_report(y_test,lr_pred))
```

```python
print(confusion_matrix(y_test,lr_pred))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
features=[columnsNotToDrop]
for feature in features:
    dataset[feature]=scaler.fit_transform(dataset[feature])
dataset[columnsNotToDrop]
```

```python
X_trainsc, X_testsc, y_trainsc, y_testsc = train_test_split(dataset[columnsNotToDrop],y, test_size=0.20, random_state=42)
```

```python
lrsc=LogisticRegression(class_weight="balanced")
lrsc.fit(X_trainsc,y_trainsc)
lrsc_pred=lrsc.predict(X_testsc)
```

```python
print(accuracy_score(y_testsc,lrsc_pred))
```

```python
print(classification_report(y_testsc,lrsc_pred))
```

```python
print(confusion_matrix(y_testsc,lrsc_pred))
```
